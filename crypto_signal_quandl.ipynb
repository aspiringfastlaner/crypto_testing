```python
# Crypto latest MAs
# BTC 10-20
# ETH 15-30
# LTC 25-40
# NEO 10-50

def crypto_ma_signal(crypto, fast = 1, slow = 10):
    df = crypto.copy()
    df['fast'] = df[['Last']].rolling(fast).mean().shift(1)
    df['slow'] = df['Last'].rolling(slow).mean().shift(1)
    
    df['return'] = df['Last']/df['Last'].shift(1) - 1
    df['signal'] = df['fast'] >= df['slow']
    
    return df[['Last','fast','slow','signal']].tail(5)
```
